```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
from mtcnn.mtcnn import MTCNN
```

```python
detector= MTCNN()
```

```python
image=cv2.imread('/Users/garima/GAN/swimteam_test.jpg')
```

```python
detector.detect_faces(image)
```

```python
cap =cv2.VideoCapture(0)
```

```python
while True:
    #capture frame by frame
    _ , frame=cap.read()
    #use MTCNN to detect faces
    faces=detector.detect_faces(frame)
    if faces != []:
        for person in faces:
            bounding_box = person['box']
            keypoints = person['keypoints']
            
            #draw box around the face
            cv2.rectangle(frame, (bounding_box[0], bounding_box[1]), (bounding_box[0]+bounding_box[2], bounding_box[1]+bounding_box[3]),(0,155,255),2)
            
            # draw circles to show keypoints :eyes, nose, mouth
            cv2.circle(frame,(keypoints['left_eye']), 2,(0,155,255),2)
            cv2.circle(frame,(keypoints['right_eye']), 2,(0,155,255),2)
            cv2.circle(frame,(keypoints['nose']), 2,(0,155,255),2)
            cv2.circle(frame,(keypoints['mouth_left']), 2,(0,155,255),2)
            cv2.circle(frame,(keypoints['mouth_right']), 2,(0,155,255),2)
            
            #display resulting frame
            plt.imshow(frame)
            if cv2.waitkey(1) &0xFF == ord('q'):
                break
            
            
cap.release()
cap.destroyAllWindows
```
